# Importing required files

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import s3fs
```

# Keeping only Restaurants business

```
business = pd.read_json("https://yelpds5230.s3.amazonaws.com/yelp_academic_dataset_business.json", lines=True)
```

```
business.head()
```

```
# Fitering open businesses
business = business[business['is_open'] == 1]
```

```
#finding total categories
trial = business.assign(categories = business.categories
                         .str.split(', ')).explode('categories')
```

```
series = trial.categories.value_counts()
```

```
series
```

```
check_series = pd.DataFrame(data = series.values, index = series.index)
```

```
check_series.head(20)
```

```
cats = ['Restaurants', 'Salad', 'Chicken Wings', 'Cafes', 'Desserts', 'Bakeries', 'Chinese', 'American (New)', 'Italian', 'Mexican', 'Breakfast & Brunch', 'Burgers', 'Pizza', 'American (Traditional)', 'Coffee & Tea', 'Sandwiches', 'Fast Food', 'Bars', 'Food', 'Restaurants']
```

```
# Keeping only restaurants/eateries
rest = trial[trial['categories'].isin(cats)]
```

```
rest.shape
```

```
trial.shape
```

```
rest.head()
```

```
only_restaurants = rest.drop_duplicates(subset='name')
```

```
only_restaurants.shape
```

```
only_restaurants.isna().sum()
```

```
# small comments
tips = pd.read_json('https://yelpds5230.s3.amazonaws.com/yelp_academic_dataset_tip.json', lines=True)
```

```
#join
yt = pd.merge(only_restaurants, tips, how='left', on='business_id')
```
