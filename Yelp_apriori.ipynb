```
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
```

```
from nltk.stem import WordNetLemmatizer 
  
lemmatizer = WordNetLemmatizer() 
```

```
array = []
def text_clean(df):
  for _, row in df.iterrows():
    text = row['text']
    text = text.lower()  #lowercase
    words = nltk.word_tokenize(text)    #tokenize
    new_words= [word for word in words if word.isalpha()]   #Remove punctuations and numbers
    nostop = [words for words in new_words if words not in ENGLISH_STOP_WORDS]    #Remove stopwords
    #nopos = [words for words in nostop if words not in positive_words] 
    lema = [lemmatizer.lemmatize(words) for words in nostop]   #lemmatize words
    array.append(lema)
  return array
```

```
words = text_clean(df)
```

```
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
```

```
trans_encode = TransactionEncoder()
```

```
encoded_array = trans_encode.fit(words).transform(words)
```

```
df_encoded = pd.DataFrame(encoded_array, columns=trans_encode.columns_)
```

```
df_encoded
```

```
df_encoded.count()
```

```
frequent_itemset = apriori(df_encoded, min_support=0.1, use_colnames=True)
```
